```python
import numpy as np 
import pandas as pd
import yfinance as yf
```

```python
tickers = [
    "AAPL","MSFT","NVDA","AMZN","GOOGL",
    "JPM","GS",
    "XOM","CVX",
    "JNJ","PG",
    "TSLA",
    "NKE","MCD"
]

data = yf.download(tickers, start = '2022-01-01', group_by = 'column')['Close']
data.head()
```

```python
initial_investment = 100000
buy_p = data.iloc[0]
buy_p
```

```python
w = buy_p/ np.sum(buy_p)
sum(w)
```

```python
shares = initial_investment * w / buy_p
shares
```

```python
sum(shares * buy_p)
```

```python
data['port_value'] = (data * shares).sum(axis=1)
data.head()
```

```python
data['port_value'].plot()
```
